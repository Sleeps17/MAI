4. Напишите функцию, которая принимает 3 трехмерных вектора $x, y, z$ и вычисляет объем пирамиды, образованной этими векторами (используйте смешанное произведение векторов)

```python
def volume(x : np.array, y : np.array, z : np.array) -> float:
    return float(abs(np.dot(z, np.cross(x, y))) / 6.0)

volume([1, 0, 0], [0, 1, 0], [0, 0, 1])
```

5. Напишите функцию, которая принимает число N и строит матрицу NxN, с нулями внутри и единицами на границе.

```python
def ones_border_matrix(n : int) -> np.ndarray:
    m = np.ones((n, n))
    m[1: -1, 1: -1] = 0
    return m

ones_border_matrix(4)
```

7. Как найти наиболее близкое значение (к заданному скаляру) в векторе?

```python
def nearest(x : float, v : np.array) -> float:
    return float(v[np.argmin(np.abs(v - x))])

nearest(4.4, np.arange(0, 9, 0.3))
```

10. Даны 2 массива Х и Y, постройте матрицу Коши $C$ ($C_{ij} = \frac{1}{x_i - y_j}$)

```python
def koshi_matrix(x : np.array, y : np.array) -> np.ndarray:
    x_t = x.reshape(-1, 1)
    return 1 / (x_t - y)

x = np.arange(2)
y = np.arange(5, 10)

koshi_matrix(x, y)
```

### 3. Визуализация

Используя библиотеки matplotlib/sns/plotly/pandas и набор данных представленный ранее визуализируйте

1. Сделайте круговую диаграмму, на которой были бы отражено число заболеваний COVID-19 по континентам

```python
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter

color_map = plt.get_cmap()
rescale = lambda x: (x - x.min()) / (x.max() - x.min())
```

```python
columns_to_group = ['continent', 'total_cases']
grouped_cases = merged[columns_to_group].groupby('continent').sum()

continent_names = grouped_cases.index
total_cases_per_continent = grouped_cases.total_cases
overall_cases = total_cases_per_continent.sum()

def format_value(value):
    computed_value = (value / 100) * overall_cases
    if computed_value >= 1e6:
        return '{:.1f}M'.format(computed_value / 1e6)
    else:
        return '{:.0f}K'.format(computed_value / 1e3)

plt.title('COVID-19 Cases Distribution by Continent')
plt.pie(total_cases_per_continent, labels=continent_names, autopct=format_value)
plt.show()
```

2. Постройте диаграмму, на которой в порядке убывания были бы отмечены число тестов на COVID-19 в России, Китае, США, Индии и Нигерии

```python
countries = ['Russia', 'China', 'United States', 'India', 'Nigeria']
columns = ['location', 'total_tests']

formater = lambda x: f'{float(x) / 1e6:f}M' if x else '0'


agg_data = merged[merged.location.isin(countries)][columns].sort_values(by='total_tests', ascending=False).fillna(0)
location = agg_data.location
tests = agg_data.total_tests

plt.title('Кол-во тестов на COVID-19')
plt.ylabel('Кол-во тестов')
bar_container = plt.bar(location, tests, color=color_map(rescale(tests)))
plt.bar_label(bar_container, fmt=formater)
plt.show()
```

3. Визуализируйте зависимость ВВП страны и случаев заболевания COVID-19

```python
merged['gdp'] = merged.gdp_per_capita * merged.population 

agg_data = merged[['location', 'gdp', 'total_cases']].sort_values(by='total_cases', ascending=False).dropna()

plt.figure(figsize=(16, 9))
plt.title('Зависимость ВВП страны и кол-во случаев заболевания COVID-19')

plt.scatter(agg_data['gdp'], agg_data['total_cases'])
plt.xlabel('ВВП')
plt.ylabel('Кол-во случае COVID-19')

trillion = lambda x, pos: f'{x / 1e12:.0f}T USD' if x else '0 USD'
plt.gca().xaxis.set_major_formatter(FuncFormatter(trillion))

plt.grid()
plt.show()
```

6. Сравните распределения кол-ва смертей на 1000 человек в Европе и Африке

```python
colums = ['location', 'mortality']

europe_dist = merged[merged.continent == 'Europe'][colums].mortality / 1000
africa_dist = merged[merged.continent == 'Africa'][colums].mortality / 1000

plt.figure(figsize=(16, 9))
plt.title('Сравнение распределения кол-ва смертей на 1000 человек в Европе и Африке')

plt.hist(europe_dist, label='Европа')
plt.hist(africa_dist, label='Африка')

plt.xlabel('Кол-во смертей на 100 человек')
plt.ylabel('Частота')
plt.grid()
plt.legend()
plt.show()
```
